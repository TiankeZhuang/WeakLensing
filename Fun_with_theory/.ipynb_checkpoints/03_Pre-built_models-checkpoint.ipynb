# HOD-style Models #

HOD-style models make predictions for galaxy samples selected to be brighter or more massive than some threshold lower bound. The fundamental quantity is $P(N_{gal}|M_{halo})$, the probability that a halo with mass $M_{halo}$ contains $N_{gal}$ galaxies in some sample. All HOD models are based on host halo only (**what does this mean?**): no connection between subhalo and satellite abundance.

````python
from halotools.empirical_models import PrebuiltSubhaloModelFactory
model = PrebuiltSubhaloModelFactory('smhm_binary_sfr')

from halotools.sim_manager import CachedHaloCatalog
halocat = CachedHaloCatalog(simname = 'bolshoi', redshift = 0, halo_finder = 'rockstar')

model.populate_mock(halocat)

sample_mask = model.mock.galaxy_table['stellar_mass'] > 1e10
gals = model.mock.galaxy_table[sample_mask]
````

## Calculate total stellar mass $M^\mathrm{tot}_*$ in each halo ##

Perform group-wise calculations with iteration over galaxy sample. In this case, we sum up stellar masses of each group's mambers at each stap of iteration.

````python
from halotools.utils import group_member_generator

gals.sort('halo_hostid')
grouping_key = 'halo_hostid'
requested_columns = ['stellar_mass']
group_gen = group_member_generator(gals, grouping_key, requested_columns)
````

What I think is happening here:

1. group_gen groups together galaxies of the same ```halo_hostid```;

2. ```first``` and ```last``` stores indices to access galaxies with the same ```halo_hostid```, while ```member_props``` stores all the values associated with the galaxy property of concern. If given more properties, the array would have more dimensions;

3. So the question is why are there repeated indices in ```host_upid```?


````python
total_stellar_mass = np.zeros(len(gals))
for first, last, member_props in group_gen:
    stellar_mass_of_members = member_props[0]
    total_stellar_mass[first:last] = sum(stellar_mass_of_members)
````
